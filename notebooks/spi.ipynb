```python
import matplotlib.pyplot as plt

import pandas as pd
import zipfile
pth ="../stars/data/us/spi.zip"
```

```python
zf = zipfile.ZipFile(pth)
```

```python
zf.namelist()
```

```python
sa1 = pd.read_csv(zf.open('SA1_1929_2017.csv'))
```

```python
sa1
```

```python
pcincome = sa1[sa1.LineCode==3.0]
```

```python
pcincome.head()
```

```python
pcincome = pcincome.drop(columns=['Table', 'LineCode', 'IndustryClassification', 'Description'])
```

```python
pcincome.head()
```

```python
m = pd.melt(pcincome, id_vars=['GeoFIPS', 'GeoName', 'Region'])
```

```python
m.head()
```

```python
m.rename(columns={'variable': 'Year', 'value': 'PC Income'}, inplace=True)
```

```python
m.head()
```

```python
m[m['PC Income']=='(NA)']=0 # fix NA's 
```

```python
m['PC Income'] = pd.to_numeric(m['PC Income'])
```

```python
cal = m[m.GeoName=='California']['PC Income'].values
```

```python
cal
```

```python
geonames = m.GeoName.unique()
```

```python
import numpy as np
geonames = m.GeoName.unique()
n,k = m.shape
uspci = m[m.GeoName=='United States']['PC Income'].values
m['relative'] = np.zeros((n,1))
geonames = geonames.tolist()
geonames.remove(0)
geonames.remove('Alaska*')
geonames.remove('Hawaii*')
for geoname in geonames:
    pci = m[m.GeoName==geoname]['PC Income'].values
    relative = pci / uspci
    m.loc[m.GeoName==geoname,'relative']= relative
    #m[m.GeoName==geoname]['relative'] = relative
    
```

```python
m.head()
```

```python
m[m.GeoName=='Alabama'].plot(x='Year', y='relative')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
for i,c in enumerate(geonames):
    m[m.GeoName==c].plot(x='Year', y='relative',ax = axes, label=c)
```

```python
summary = pd.DataFrame()
```

```python
summary['GeoName'] = geonames
```

```python
summary['gr'] = np.zeros((len(geonames),1))
summary['initial'] = np.zeros((len(geonames),1))
summary['ending'] = np.zeros((len(geonames),1))
```

```python
summary.shape
```

```python
summary.head()
```

```python
for geoname in geonames:
    vals = m.loc[m.GeoName==geoname, 'relative'].values
    summary.loc[summary.GeoName==geoname,'gr'] = vals[-1]/vals[0]
    summary.loc[summary.GeoName==geoname,'initial'] = vals[0]
    summary.loc[summary.GeoName==geoname,'ending'] = vals[-1]
```

```python
from matplotlib.colors import LinearSegmentedColormap
a = 0.7
n = len(df)
customcmap =  [(x/n,  x/(n+n), 0.05) for x in range(n)]

fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111)
df.plot.barh(x='GeoName',y='gr',
             xlim=(0, max(df['gr'])),
             color=customcmap,
             ax=ax, title=ttl)

ax.grid(False)
ax.set_frame_on(False)
ax.set_title(ax.get_title(), fontsize=26, alpha=a, ha='left')
plt.subplots_adjust(top=0.9)
ax.title.set_position((0,1.08))
 
# Set x axis label on top of plot, set label text
ax.xaxis.set_label_position('top')
xlab = 'Relative Income Growth'
ax.set_xlabel(xlab, fontsize=20, alpha=a, ha='left')
ax.xaxis.set_label_coords(0, 1.04)
 

# Position x tick labels on top
ax.xaxis.tick_top()
# Remove tick lines in x and y axes
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_ticks_position('none')

# Customize x tick lables
xticks = [.25, .50, .75, 1.0, 1.25, 1.5, 1.75, 2.0]
ax.xaxis.set_ticks(xticks)
ax.set_xticklabels(xticks, fontsize=16, alpha=a)
 
    
yticks = [item.get_text() for item in ax.get_yticklabels()]
ax.set_yticklabels(yticks, fontsize=10, alpha=a)
ax.yaxis.set_tick_params(pad=12)  


# Create a fake colorbar
ctb = LinearSegmentedColormap.from_list('custombar', customcmap, N=2048)
# Trick from http://stackoverflow.com/questions/8342549/
# matplotlib-add-colorbar-to-a-sequence-of-line-plots
#sm = plt.cm.ScalarMappable(cmap=ctb, norm=plt.normalize(vmin=72, vmax=84))
sm = plt.cm.ScalarMappable(cmap=ctb)

# Fake up the array of the scalar mappable
sm._A = []
 
# Set colorbar, aspect ratio
cbar = plt.colorbar(sm, alpha=0.05, aspect=16, shrink=0.4)
cbar.solids.set_edgecolor("face")
# Remove colorbar container frame
cbar.outline.set_visible(False)
# Fontsize for colorbar ticklabels
cbar.ax.tick_params(labelsize=16)
# Customize colorbar tick labels
mini = df.initial.min()
maxi = df.initial.max()
mytks = np.linspace(mini,maxi, 5).tolist()

#cbar = fig.colorbar(sm, ticks=[mini, 1.0, maxi])
#cbar.ax.set_yticklabels(['0.38', '1.0', '1.89']) 
#mytks = range(mini, maxi, 0.1)
rg = maxi - mini
sticks = [ (y-mini) / rg for y in mytks]
#cbar.set_ticks(mytks)
cbar.set_ticks(sticks)

#cbar.ax.set_yticklabels([str(a) for a in mytks], alpha=a)
cbar.ax.set_yticklabels(["%1.2f"%a for a in mytks], alpha=a)
# Colorbar label, customize fontsize and distance to colorbar
cbar.set_label('Initial Relative Incomes (US=1.0)', alpha=a, 
               rotation=270, fontsize=20, labelpad=20)
# Remove color bar tick lines, while keeping the tick labels
cbarytks = plt.getp(cbar.ax.axes, 'yticklines')
plt.setp(cbarytks, visible=False)
```
